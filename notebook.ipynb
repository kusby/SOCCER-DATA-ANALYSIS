# 1. What team commits the most fouls?

```python
# Home teams' most fouls commited

home_team_foul = epl_df.pivot_table(index=["HomeTeam"], values=["HF"], aggfunc="sum").sort_values(by='HF', ascending=False).reset_index()
home_team_foul

```

```python
# Visualizing the most fouls committed by Home team
plt.figure(figsize=(20,6))
plt.title("Numbers of Foul Committed By Home Team", fontsize=20)
sns.barplot(data = home_team_foul, x = 'HomeTeam', y = 'HF')
plt.ylabel("Fouls Committed", fontsize=15)
plt.xlabel("Home Team", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# Away teams' most fouls commited

away_team_foul = epl_df.pivot_table(index=["AwayTeam"], values=["AF"], aggfunc="sum").sort_values(by='AF', ascending=False).reset_index()
away_team_foul
```

```python
# Visualizing the most fouls committed by Away team
plt.figure(figsize=(20,6))
plt.title("Numbers of Foul Committed By Away Team", fontsize=20)
sns.barplot(data = away_team_foul, x = 'AwayTeam', y = 'AF')
plt.ylabel("Fouls Committed", fontsize=15)
plt.xlabel("Away Team", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# What team commits the most most fouls?

hometeamfoul = epl_df[['HomeTeam', 'HF']].groupby('HomeTeam').sum()
awayteamfoul = epl_df[['AwayTeam', 'AF']].groupby('AwayTeam').sum()

team_fouls = hometeamfoul['HF'] + awayteamfoul['AF']

team_fouls = pd.DataFrame(team_fouls)

team_fouls = team_fouls.reset_index().sort_values(0, ascending = False)

team_fouls = team_fouls.rename(columns = {'HomeTeam' : 'Team'})

team_fouls


```

```python
# Visualizing the most fouls committed by Away team
plt.figure(figsize=(20,6))
plt.title("Numbers of foul committed By each team", fontsize=20)
sns.barplot(x = team_fouls.Team, y = team_fouls[0].values)
plt.ylabel("Fouls Committed", fontsize=15)
plt.xlabel("Football Team", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

# Report

. With 463 fouls, Brighton committed most fouls in the 2018/2019 EPL season while Liverpool committed the least which stood at 315 fouls.


# Plot the percentage of games that ended in a draw over time

```python
# Plot the percentage of games that ended in a draw over time.
games_draw_pct = epl_df["FTR"].value_counts(normalize=True)

games_draw_pct


    
```

```python
# Visualizing the percentage of games that ended in a draw over time.
plt.figure(figsize=(10,3))
games_draw_pct.plot()
plt.xlabel("Full Time Result")
plt.ylabel("Percentage")
plt.title("Percentage of Home Win(H), Away Win(A), & Draw(D)", fontsize=10)
plt.show()
```

## Report

. The plot above shows 18% of the games played in the 2018/2019 EPL season resulted in a draw.

# Does the number of red cards a team receives have an effect on its probability of winning a game?

```python
# Home team red card
home_red = epl_df.pivot_table(index=["HomeTeam"], values=["HR"], aggfunc="sum").sort_values(by="HR", ascending=False).reset_index()

home_red
```

```python
# Away team red card
away_red = epl_df.pivot_table(index=["AwayTeam"], values=["AR"], aggfunc="sum").sort_values(by="AR", ascending=False).reset_index()

away_red
```

```python
# Team with most red cards

homeredcards = epl_df[['HomeTeam', 'HR']].groupby('HomeTeam').sum()

awayredcards = epl_df[['AwayTeam', 'AR']].groupby('AwayTeam').sum()

team_red_card = homeredcards['HR'] + awayredcards['AR']

team_red_card = pd.DataFrame(team_red_card)

team_red_card = team_red_card.reset_index().sort_values(0, ascending = False)

team_red_card = team_red_card.rename(columns = {'HomeTeam' : 'Team'})

team_red_card
```

```python
# Visualizing team with highest number of red cards
plt.figure(figsize=(20, 6))
sns.barplot(x = team_red_card.Team, y = team_red_card[0].values, data = team_red_card)
plt.title("Numbers of red cards received by each team", fontsize=20)
plt.ylabel("Numbers of red cards", fontsize=15)
plt.xlabel("Football team", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# Create Home Win, Draw,Away Win and RedCard columns. 
epl_df_2 = epl_df.assign(HomeWin=lambda epl_df: epl_df.apply(lambda x: 1 if x.FTHG > x.FTAG else 0,
                         axis='columns'),
                         
             			 Draw=lambda epl_df: epl_df.apply(lambda x: 1 if x.FTHG == x.FTAG else 0,
                         axis='columns'),
                         
              			 AwayWin=lambda epl_df: epl_df.apply(lambda x: 1 if x.FTHG < x.FTAG else 0,
                         axis='columns'),
                         
              			 RedCard=lambda epl_df: epl_df.apply(lambda x: 'BothRed' if (x.HR >= 1 and x.AR >= 1) else
                         'HomeRed' if x.HR >= 1 else 'AwayRed' if x.AR >= 1  else 'NoRed', axis='columns'))


win_pct = epl_df_2.groupby('RedCard').mean().loc[:, ["HomeWin", "Draw", "AwayWin"]]

win_pct
```
